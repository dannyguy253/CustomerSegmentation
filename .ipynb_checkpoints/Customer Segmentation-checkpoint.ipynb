# Python for Data Cleaning and Preprocessing

Objective: Clean and preprocess the data to prepare it for clustering.

We will now use the exported file from using SQL to do K-Means Clustering

````python
import pandas as pd
df = pd.read_csv('customer_segmentation.csv')
````

````python
# No duplicates
df.duplicated().sum()
````

````python
# No nulls
df.isnull().sum()
````

````python
# Convert dtypes of the purchase date cols to datetime
df['First_Purchase_Date'] = pd.to_datetime(df['First_Purchase_Date'])
df['Last_Purchase_Date'] = pd.to_datetime(df['Last_Purchase_Date'])
````

````python
df.info()
````

````python
# Feature Scaling to standardize features to have mean 0 and variance 1
````

````python
from sklearn.preprocessing import StandardScaler

features = ['Num_Visits', 'Total_Spent','recency', 'Avg_Spend', 'Avg_Items', 'Discount_Usage']
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])
````

````python
df
````

# Python for Clustering Algorithm

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# List to store inertia values for each k
inertia = []

# Range of k values to try (from 1 to 10)
K = range(1, 11)

# Fit KMeans for each k and store the inertia
for k in K:
    # n_init=10 means to run the algorithm 10 times for stability and choose lowest inertia of the 10
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df[features])
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
````

To find the elbow (aka optimal k) we look for the point where the curve bends or elbows. This should be the point where adding another cluster does not significantly reduce the inertia. So, the most noticeable "elbow" appears at k=3. 

````python
# Fit K Means
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(df[features])
````

# Python: Visualization

Objective: Visualize the resulting clusters.

Pair Plot: Use Seaborn to visualize the clusters.

Each subplot in the pair plot shows the scatter plot between two features, allowing you to visually assess the separation between clusters.

Main diagonal plots are individual variable distributions and off diagonal plots are scatter plots between any two variables. All plots are color coded by each cluster. 

````python
import seaborn as sns

sns.pairplot(df, hue='cluster', vars=features, palette='Set2')
plt.show()
````

Centroid Visualization: Plot centroids to understand cluster centers

````python
# centers of each of the k=4 clusters
# inverse transform transforms the centroids back into the original feature space to help make the values interpretable
cluster_centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=features)
cluster_centers
````

Summary:
Customers in Cluster 2 are infrequent visitors with the lowest total and average spending per visit. They purchase a similar number of items per visit and have low discount usage, similar to Cluster 1.
Overall Insights:

    Cluster 0: Moderate visit frequency, moderate total and average spend, higher discount usage.
    Cluster 1: Infrequent visits, high average spend per visit, moderate discount usage.
    Cluster 2: Infrequent visits, low total and average spend, moderate discount usage.

### 3. **Python: Data Cleaning and Preprocessing**

**Objective:** Clean and preprocess the data to prepare it for clustering.

- **Steps:**
  - **Load Data:** Import the data into Python using Pandas.
  - **Data Cleaning:** Handle missing values, correct data types, and manage outliers.
    ```python
    import pandas as pd

    # Load data
    data = pd.read_csv('customer_data.csv')

    # Handle missing values
    data.fillna(0, inplace=True)

    # Convert date columns to datetime
    data['last_purchase_date'] = pd.to_datetime(data['last_purchase_date'])

    # Calculate recency in days
    data['recency'] = (pd.Timestamp.now() - data['last_purchase_date']).dt.days
    ```
  - **Feature Scaling:** Standardize features to have a mean of 0 and a standard deviation of 1.
    ```python
    from sklearn.preprocessing import StandardScaler

    features = ['frequency', 'total_spent', 'recency']
    scaler = StandardScaler()
    data[features] = scaler.fit_transform(data[features])
    ```

### 4. **Python: Clustering**

**Objective:** Apply clustering algorithms to segment customers.

- **Steps:**
  - **Choose Clustering Algorithm:** Use K-means for simplicity and interpretability.
  - **Determine Optimal Number of Clusters:** Use the Elbow method or Silhouette analysis to find the optimal number of clusters.
    ```python
    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt

    # Determine the optimal number of clusters
    inertia = []
    K = range(1, 11)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data[features])
        inertia.append(kmeans.inertia_)

    # Plot the Elbow curve
    plt.figure(figsize=(10, 6))
    plt.plot(K, inertia, 'bx-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal K')
    plt.show()
    ```
  - **Fit K-means:**
    ```python
    optimal_k = 3  # Replace with the optimal number determined
    kmeans = KMeans(n_clusters=optimal_k, random_state=42)
    data['cluster'] = kmeans.fit_predict(data[features])
    ```
